```python
#2.ML Pipeline - Preprocessing
X = youtube.drop('Total_Energy(kJ)',axis=1)
#ID,title,Channel_title not important
X = X.drop(['ID','Title','Channel_title'],axis=1)
y = youtube['Total_Energy(kJ)']

#split to train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)

#split matrix into numerical and categorical attributes
X_train_num=X_train.drop(['Country'],axis=1)
X_train_cat=X_train['Country']

#Found missing value
print('Number of missing values for all columns:\n',X_train_num.isnull().sum())
X_train_num=X_train_num.fillna(0)
print('nAfter handling Missing values:\n',X_train_num.isnull().sum())

#Standardize numeric data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler(copy = False)
scaler.fit(X_train_num)
X_train_num_tr = scaler.transform(X_train_num)

#One-hot encode categorical data
from sklearn.preprocessing import LabelBinarizer
lb_encoder = LabelBinarizer()
lb_encoder.fit(X_train_cat)
X_train_cat_tr = lb_encoder.transform(X_train_cat) # returns numpy array

#combine the transformed numerical and categorical sets
X_train_tr = np.hstack([X_train_num_tr,X_train_cat_tr])
y_train = y_train.values; #convert to numpy array
print('\nX_train_tr shape:',X_train_tr.shape)
print('y_train shape:',y_train.shape)
```

```python
#3.ML Pipeline - Model selection and building the model
#Linear Regression model
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error

# create a instance of LinearRegression
lin_reg = LinearRegression()

# Train the regression model
lin_reg.fit(X_train_tr, y_train)

# Predict on training set using trained regression
y_pred =  lin_reg.predict(X_train_tr)

# Function to choose 10 random samples
def show10results(y_train,y_pred):
    print('Result for the 10 random samples:') 
    selected = np.random.randint(0, len(y_train), 10) 
    for i in selected:
        print('actual = {:7.0f} pred ={:7.0f}'.format(y_train[i], y_pred[i])) 

show10results(y_train, y_pred)
# Compute and show the RMSE on training set 
lin_mse = mean_squared_error(y_pred, y_train) 
lin_rmse = np.sqrt(lin_mse)
print ('RMSE =', lin_rmse) #Underfitting
print ('mean=',youtube['Total_Energy(kJ)'].mean ())
```

```python
from sklearn .tree import DecisionTreeRegressor

# create a instance of DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(random_state=42)

# Train the regressor model
tree_reg.fit (X_train_tr, y_train)

# Predict on training set using trained regressor
y_pred =  tree_reg.predict(X_train_tr)

# Show the result of 10 random samples
show10results(y_train, y_pred)

#Compute ond  show the RMSE  on training  set 
tree_mse = mean_squared_error(y_pred, y_train) 
tree_rmse = np.sqrt(tree_mse)
print('RMSE =', tree_rmse) #Overfitting
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=10, random_state=42) 
forest_reg.fit (X_train_tr, y_train)

# Predict on troining set
y_pred = forest_reg.predict(X_train_tr)

#Show result of 10 random samples
show10results(y_pred, y_train)

# Show the RMSE
forest_mse = mean_squared_error(y_pred, y_train) 
forest_rmse = np.sqrt(forest_mse)
print('RMSE =', forest_rmse)#Forest Regressor will be fine on predicting model
```

```python
#4.ML Pipeline - Model Validation
#linear Regression Model

from sklearn.model_selection import cross_val_score

#cross validation - 3 fold
k_fold_scores = cross_val_score(lin_reg, X_train_tr, y_train, scoring = 'neg_mean_squared_error', cv=3)

linreg_rmse_scores = np.sqrt(-k_fold_scores)

#function to display scores,mean & standard deviation
def display_scores(scores): 
    print('Scores:', scores,'\n') 
    print('Mean:', scores.mean())
    print('Standard deviation:', scores.std())

display_scores(linreg_rmse_scores)
```

```python
def show_samples (X, y, axis=[0,15000000,0,1500000]): 
    fig = plt.figure(figsize=(8, 6))
    plt.plot(X, y, 'b.', markersize=12,label='samples')
    plt.xlabel ('$TotalEnergy(kJ)$', fontsize=18) 
    plt.ylabel('$Views$', fontsize=18)
    plt.axis(axis)
    plt.legend(loc='best')

# Visualize the result j
Xtest =  np.linspace(0, 15000000,100).reshape(-1,1)
ypred = lin_reg.predict(Xtest) 
show_samples(X, y)
plt.plot (Xtest, ypred, 'r-', label = 'model (LinearRegression) ')
plt.legend() 
plt.show()
```

```python
class MyLinearRegression :

    def __init__(self, theta = None): 
        self.theta = np.array(theta)
        
    def fit(self, X, y):
        m, n = X.shape
        Xo= np.hstack((np.ones((m,1)),X))
        self.theta =np.linalg.pinv(Xo.T.dot(Xo)).dot(Xo.T.dot(y))
        
    def predict(self, X):
        m,n=X.shape
        Xo=np.hstack((np.ones((m,1)),X)) 
        h=Xo.dot(self.theta)
        return h
```

```python
#Cost function - Evaluate  how good is your model compare with other theta
show_samples(X, y)

#good model
#thetha  is  calculated from  normal  equation
#y = 100000x + 0.09
model1 = MyLinearRegression(theta = [100000, 0.09]) 
h1 = model1.predict(X)
plt.plot(X, h1, 'r-', label = 'model1 (theta = [100000,0.09])')

#bad  model
model2 = MyLinearRegression(theta = [4,0.3]) 
h2 = model2.predict(X)
plt.plot(X, h2, 'y-', label = 'model2 (theta = [4, 0.3])')

plt.legend() 
plt.show()
```

```python
def RMSE(y_true, y_pred):
    rmse=np.sqrt(np.mean((y_true-y_pred)**2))
    return rmse

print('RMSE for model1 = ',RMSE(y,h1))
print('RMSE for model2 = ',RMSE(y,h2))
#Model 1 fits data than model 2
```

```python
# Implement the training module
# Find optimal
mylinreg = MyLinearRegression()
mylinreg.fit(X, y)
print('Fitted theta:',mylinreg.theta)

# Show the learnt model
h_pred = mylinreg.predict(X)
print('Training error (RMSE):', RMSE(y, h_pred))

# Visualize the result
Xtest = np.linspace(0, 15000000, 100).reshape(-1, 1) 
h_pred = mylinreg.predict(Xtest)
show_samples(X, y)
plt.plot (Xtest, h_pred, 'r-', label = 'model (MyLinearRegression)') 
plt.legend()
plt.show()
```
